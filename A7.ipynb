```python
from nltk import word_tokenize, sent_tokenize
from nltk import pos_tag
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
```

```python
sentence = "Sachin was the GOAT of the previous generation. Virat is the GOAT of this generation. Shubman will be the GOAT of the next generation"
```

```python
#Tokenization
print('Tokenized words:', word_tokenize(sentence))
```

```python
print('\nTokenized sentences:', sent_tokenize(sentence))
```

```python
#POS Tagging
token = word_tokenize(sentence)
tagged = pos_tag(token)
print("Tagging Parts of Speech:", tagged)
```

```python
#Stop words
stop_words = stopwords.words('english')

token = word_tokenize(sentence)
cleaned_token = []

for word in token:
    if word not in stop_words:
        cleaned_token.append(word)
        
print('Unclean version:', token)
print('\nCleaned version:', cleaned_token)
```

```python
#Stemmming
stemmer = PorterStemmer()
token = word_tokenize(sentence)

stemmed = [stemmer.stem(word) 
           
for word in token]
print(" ".join(stemmed))
```

```python
#Lemmatization
lemmatizer = WordNetLemmatizer()

token = word_tokenize(sentence)

lemmatized_output = [lemmatizer.lemmatize(word) for word in token]
print(" ".join(lemmatized_output))
```
